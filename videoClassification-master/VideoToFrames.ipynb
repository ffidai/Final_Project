```python
import cv2
import numpy as np
import os
```

```python
cap = cv2.VideoCapture('LeftForehand.MOV')

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print ('Error: Creating directory of data')

```

```python
currentFrame = 0
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    #print("frame" , np.expand_dims(frame,axis=0))
    
    if not ret:
        break

    # Saves image of the current frame in jpg file
    name = './data/forehand/frame' + str(currentFrame) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)
    cv2.waitKey(200)
    # To stop duplicate images
    currentFrame += 1
```
